```python
import pandas as pd

menu = pd.read_csv("/Volumes/workspace/pandas/pandas/menu_items.csv")
orders = pd.read_csv("/Volumes/workspace/pandas/pandas/order_details.csv")

# Correct date parsing (2-digit year)
orders['order_date'] = pd.to_datetime(
    orders['order_date'],
    format='%m/%d/%y'
)

# Correct merge
combined = orders.merge(
    menu,
    left_on='item_id',
    right_on='menu_item_id',
    how='left'
)

# Orders by weekday
orders_by_day = (
    combined
    .groupby(combined['order_date'].dt.day_name())
    .size()
)

print(orders_by_day)
```

Privacy Concerns in Restaurant Data

When analyzing restaurant data, key privacy risks arise from collecting or linking personally identifiable information (PII), such as:

Names
Phone numbers
Email addresses
Credit/debit card data
Device/location tracking

These can be used to track a customer’s behavior across visits, menus, or even locations — without their full awareness or consent. If breached or misused, this could expose sensitive consumer habits or financial information.

Best Practice: Anonymize or pseudonymize PII (remove or mask identifiers), and enforce strict access control to prevent internal misuse or external breaches.

Fairness Concerns

Fairness issues can arise when customer data is used:

To dynamically change prices (price discrimination)

To target certain users unfairly (e.g., upsell more to lower-income users based on behavior)

Or when users are profiled or tracked without consent

For example, if a customer is unknowingly tracked and targeted with personalized pricing or recommendations based on spending habits, it can feel manipulative or even discriminatory — especially if there is no transparency or ability to opt out.

Best Practice: Ensure informed consent for data collection, provide opt-out mechanisms, and audit algorithms for biased behavior.

Using Data to Serve (Not Manipulate) Customers

To ensure customer data serves rather than manipulates: Transparency: Be upfront about what data is collected and how it’s used. Use plain language in privacy policies. Consent: Always collect opt-in permission before tracking behavior. Limitations on Use: Use data to improve experience (e.g., faster service, favorite orders) — not to exploit spending habits or manipulate decisions. Ethical Algorithms: Regularly audit ML/AI models to ensure they’re not reinforcing bias (e.g., recommending more expensive items unfairly). Privacy by Design: Build systems to protect user data from the start (e.g., encryption, access logging, minimal data retention).

A good rule: If the customer knew how you used their data, would they still trust you?

To protect privacy and promote fairness, we must anonymize data, limit access, and be transparent about usage. Responsible data practices help ensure insights are used to improve customer experiences — not manipulate choices or invade trust.
