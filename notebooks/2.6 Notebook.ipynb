## Bronze Table Summaries

### device_message_raw
This table stores raw sensor readings generated by STEDI devices, including timestamps, sensor type, and distance measurements. The data captures individual sensor events over time and serves as the primary source of signal data for modeling movement patterns. Understanding these fields is essential for cleaning and labeling sensor data in the Silver layer.

### rapid_step_test_raw
This table records metadata about rapid step test sessions, including start and stop times, total steps, and step point scores. It defines the time windows during which valid step activity occurs, which is critical for labeling sensor readings as step or no-step. This table provides ground truth for supervised learning.


Step 1 Completed

```python
%sql
CREATE SCHEMA IF NOT EXISTS workspace.silver;

CREATE TABLE IF NOT EXISTS workspace.silver.labeled_step_test (
  -- Lineage and identifiers
  timestamp TIMESTAMP COMMENT
    'Exact time when the sensor reading was recorded. Used to trace data back to bronze.',
  device_id STRING COMMENT
    'Device identifier shared across sensor readings and step tests.',
  
  -- Raw sensor fields
  sensor_type STRING COMMENT
    'Type of sensor that produced the reading.',
  distance_cm INT COMMENT
    'Clean numeric distance value converted from string values like "75cm".',
  
  -- Labels created during ETL
  step_label STRING COMMENT
    'Label indicating whether the reading occurred during a step test window ("step" or "no_step").',
  source STRING COMMENT
    'Indicates the original data source (e.g., device_message).',
  
  -- Step test window metadata
  start_time TIMESTAMP COMMENT
    'Start time of the rapid step test window.',
  stop_time TIMESTAMP COMMENT
    'End time of the rapid step test window.'
)
USING DELTA;
```

Step 2 Completed

```python
%sql
SELECT 
  column_name AS `Column Name`,
  data_type AS `Data Type`,
  comment AS `Comment`
FROM workspace.information_schema.columns
WHERE table_schema = 'silver'
  AND table_name = 'labeled_step_test';
```
